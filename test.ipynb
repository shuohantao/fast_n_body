```python
center = [0.5, 0.5, 0.5]
half_size = 0.5
```

```python
children = []
for i in range(8):
    i = i+1
    child_center = [
        center[0] + (i % 2) * half_size/2,
        center[1] + ((i // 2) % 2) * half_size/2,
        center[2] + ((i // 4) % 2) * half_size/2
    ]
    children.append(child_center)
x = [child[0] for child in children]
y = [child[1] for child in children]
z = [child[2] for child in children]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, cmap='jet')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
